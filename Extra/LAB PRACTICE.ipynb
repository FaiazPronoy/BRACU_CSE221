```python
import math
import heapq

input = open("input.txt")
output = open("output.txt", "w")

vert, edges = [int(x) for x in input.readline().strip().split(" ")]
dict1 = {}
string = ['Kentucky', 'Texus', 'Alabama', 'Washington', 'Kansus', 'Delaware', 'California', 'Florida']

# Create a mapping from names to indices
name_to_index = {name: index for index, name in enumerate(string)}

for i in range(0, vert):
  dict1[string[i]] = []

start = 0
src_lst = []
while start < edges:
  source, dest, cost = input.readline().strip().split(" ")
  src_lst.append(source)
  dict1[source].append([dest, int(cost)])
  start += 1

def dijkstara(graph, initial):
  distance = [math.inf]*(vert+1)
  distance[name_to_index[initial]] = 0

  queue= []
  visited = []
  queue.append([initial, 0])
  visited.append(initial)

  while queue != []:
    heapq.heapify(queue)
    vertx, cost = heapq.heappop(queue)

    for adj_vert, dis in graph[vertx]:
      if dis + distance[name_to_index[vertx]] < distance[name_to_index[adj_vert]]:
        distance[name_to_index[adj_vert]] = dis + distance[name_to_index[vertx]]
        heapq.heappush(queue, [adj_vert, distance[name_to_index[adj_vert]]])
        visited.append(adj_vert)
    return visited, distance

visited, distance = dijkstara(dict1, src_lst[0])
print(distance)
print(visited)
```

```python
# A Huffman Tree Node
import heapq


class node:
    def __init__(self, freq, symbol, left=None, right=None):
        # frequency of symbol
        self.freq = freq

        # symbol name (character)
        self.symbol = symbol

        # node left of current node
        self.left = left

        # node right of current node
        self.right = right

        # tree direction (0/1)
        self.huff = ''

    def __lt__(self, nxt):
        return self.freq < nxt.freq


# utility function to print huffman
# codes for all symbols in the newly
# created Huffman tree
def printNodes(node, val=''):

    # huffman code for current node
    newVal = val + str(node.huff)

    # if node is not an edge node
    # then traverse inside it
    if(node.left):
        printNodes(node.left, newVal)
    if(node.right):
        printNodes(node.right, newVal)

        # if node is edge node then
        # display its huffman code
    if(not node.left and not node.right):
        print(f"{node.symbol} -> {newVal}")


# characters for huffman tree
chars = ['A', 'B', 'C', 'D', '#']

# frequency of characters
freq = [35, 15, 25, 10, 20]

# list containing unused nodes
nodes = []

# converting characters and frequencies
# into huffman tree nodes
for x in range(len(chars)):
    heapq.heappush(nodes, node(freq[x], chars[x]))

while len(nodes) > 1:

    # sort all the nodes in ascending order
    # based on their frequency
    left = heapq.heappop(nodes)
    right = heapq.heappop(nodes)

    # assign directional value to these nodes
    left.huff = 0
    right.huff = 1

    # combine the 2 smallest nodes to create
    # new node as their parent
    newNode = node(left.freq+right.freq, left.symbol+right.symbol, left, right)

    heapq.heappush(nodes, newNode)

# Huffman Tree is ready!
printNodes(nodes[0])
```
