```python
#TASK_1a
import numpy as np
input = open("input1a.txt")
output = open("output1a.txt", "w")

vert, edges = [int(x) for x in input.readline().strip().split(" ")]
row = vert + 1
col = vert + 1
matrix = np.zeros((row, col), dtype = int)

start = 0
while start < edges:
  source, dest, cost = [int(x) for x in input.readline().strip().split(" ")]
  for i in range(row):
    for j in range(col):
      if i == source and j == dest:
        matrix[i][j] = cost
  start += 1

for i in range(row):
  for j in range(col):
    output.write(f"{matrix[i][j]} ")
  output.write("\n")
output.close()
```

```python
#TASK_2
input = open("input2.txt")
output = open("output2.txt", "w")

vert, edges = [int(x) for x in input.readline().strip().split(" ")]
dict1 = {}
for i in range(1, vert+1):
  dict1[i] = []
start = 0
while start < edges:
  source, dest = [int(x) for x in input.readline().strip().split(" ")]
  dict1[source].append(dest)
  dict1[dest].append(source)
  start += 1

def bfs(graph):
  queue = []
  visited = []
  result = []
  initial = 1
  queue.append(initial)
  visited.append(initial)

  while queue != []:
    explore_vert = queue.pop(0)
    result.append(explore_vert)
    for adj_vert in graph[explore_vert]:
      if adj_vert not in visited:
        queue.append(adj_vert)
        visited.append(adj_vert)
  return result

result = bfs(dict1)
for i in result:
  output.write(f"{i} ")

output.close()
```

```python
#TASK_4
input = open("input4.txt")
output = open("output4.txt", "w")

vert, edges = [int(x) for x in input.readline().strip().split(" ")]
dict1 = {}
for i in range(1, vert+1):
  dict1[i] = []

start = 0

while start < edges:
  source, dest = [int(x) for x in input.readline().strip().split()]
  dict1[source].append(dest)
  start+= 1

visited = []
visiting = []

def check_cycle(graph, vert):
  global visited
  global visiting

  if vert in visiting:
    return True
  elif vert in visited:
    return False
  else:
    visiting.append(vert)
    for adj_vert in graph[vert]:
      if check_cycle(graph, adj_vert) == True:
        return True
    visiting.remove(vert)
    visited.append(vert)
  return False


result = check_cycle(dict1, 1)
print(result)
```
