```python
f = open("/content/Marks Input.txt")
row = 0
while True:
  line = f.readline()
  row += 1
  if not line:
    break
  else:
    v1 = line.split(" ")[0]
    v2 = line.split(" ")[1]
    v3 = line.split(" ")[2]
    print("for row", row, "Student has scored", v1, v2, v3)
```

```python
inp = open("/content/input1b.txt")
out = open("/content/output1b.txt", "a")

len = int(inp.readline())
for idx in range(len):
  line = inp.readline()
  var1, var2, var3, var4 = line.split(" ")
  if var3 == "+":
    result = int(var2) + int(var4)
    out.write(f"The result of {var2} {var3} {var4} is {result} \n")
  elif var3 == "-":
    result = int(var2) - int(var4)
    out.write(f"The result of {var2} {var3} {var4} is {result} \n")
  elif var3 == "*":
    result = int(var2) * int(var4)
    out.write(f"The result of {var2} {var3} {var4} is {result} \n")
  elif var3 == "/":
    result = int(var2) / int(var4)
    out.write(f"The result of {var2} {var3} {var4} is {result} \n")

out.close()
```
