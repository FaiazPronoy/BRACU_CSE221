```python
def factorial(n):
   if n == 0 or n == 1:
     return 1
   else:
     return n * factorial(n-1)

def helper(n, r):
        return factorial(n)//(factorial(n-r) * factorial(r))

n = 5
r = 2
print(helper(n,r))

```

```python
def ternary_search(arr, l, r, key):
    while l <= r:
        mid1 = l + (r - l) // 3
        mid2 = r - (r - l) // 3

        if arr[mid1] == key:
            return mid1
        elif arr[mid2] == key:
            return mid2

        elif key < arr[mid1]:
            r = mid1 - 1
        elif key > arr[mid2]:
            l = mid2 + 1
        else:
            l = mid1 + 1
            r = mid2 - 1


# Example usage:
arr = [1, 3, 5, 7, 9, 11, 13, 15, 17]
key = 13
result = ternary_search(arr, 0, len(arr) - 1, key)
print(result)
```

```python
def find_max_in_wave_like_array(array):
  """Finds the maximum number in a wave-like array.

  Args:
    array: A list of distinct integers in a wave-like sequence.

  Returns:
    The maximum number in the array.
  """

  # Initialize the start and end pointers.
  start = 0
  end = len(array) - 1

  # While the start and end pointers have not converged, compare the two elements
  # at the start and end pointers and update the pointers accordingly.
  while start < end:
    if array[start] > array[end]:
      return array[start]
    else:
      start += 1
      end -= 1

  # If the start and end pointers have converged, return the element at the start
  # pointer, which is the maximum element in the array.
  return array[start]

arr = [1, 3, 4, 5, 9, 6, 2, -1]
max = find_max_in_wave_like_array(arr)
print(max)
```

```python
def combination(n, r):
    if n < r:
        return 0
    elif r == 0 or n == 1:
        return 1
    elif r == 1:
        return n
    else:
        return combination(n - 1, r - 1) + combination(n - 1, r)


n = 5
r = 2
print(combination(n, r))
```

```python
# Define a function to sort the activities by their end times
def sort_by_end_time(activities):
    return sorted(activities, key=lambda x: x[1])

# Define a function to find the maximum number of activities that can be completed
def max_activities(activities, people):
    # Sort the activities by their end times
    activities = sort_by_end_time(activities)
    # Initialize the count of completed activities and the last end time
    count = 0
    last_end = 0
    # Loop through the activities
    for start, end in activities:
        # If the start time is greater than or equal to the last end time and there are people available
        if start >= last_end and people > 0:
            # Increment the count of completed activities and update the last end time
            count += 1
            last_end = end
            # Decrement the number of people
            people -= 1
    # Return the count of completed activities
    return count

# Read the input for the number of activities and people
N, M = map(int, [(1, 5), (3, 6), (2, 5), (8, 10), (6, 9)])
activities = []
# Read the start and end times for each activity
for _ in range(N):
    Si, Ei = map(int, input().split())
    activities.append((Si, Ei))

# Print the output, which is the maximum number of activities that can be completed by M people
print(max_activities(activities, M))
```

```python
def greedy_task_scheduling(tasks, people):

  tasks.sort(key=lambda task: task[1])

  people = [None] * tasks[0][1]

  # Iterate over the tasks in sorted order.
  num_completed_tasks = 0
  for task in tasks:
    # Find the first person in people who is available at the start time of the task.
    available_person = None
    for i in range(len(people)):
      if people[i] is None or people[i][1] <= task[0]:
        available_person = i
        break

    # If such a person is found, assign the task to them and set their availability to False.
    if available_person is not None:
      people[available_person] = task
      num_completed_tasks += 1

  # Return the number of tasks that were assigned to people.
  return num_completed_tasks

# Example 1
tasks = [(1, 5), (3, 6), (2, 5), (8, 10), (6, 9)]
num_completed_tasks = greedy_task_scheduling(tasks)
print(num_completed_tasks)  # 4

# Example 2
tasks = [(5,2), (1, 4), (2, 5), (6, 7), (4, 8), (3, 6)]

num_completed_tasks = greedy_task_scheduling(tasks)
print(num_completed_tasks)  # 4

# Example 3
tasks = [(1, 10), (2, 10), (6, 7), (4, 8), (3, 6)]
num_completed_tasks = greedy_task_scheduling(tasks)
print(num_completed_tasks)  # 3

# Example 4
tasks = [(5, 7), (2, 4), (6, 8), (8, 10), (1, 3), (7, 9), (3, 5), (2, 6)]
num_completed_tasks = greedy_task_scheduling(tasks)
print(num_completed_tasks)  # 8
```
