```python
#TASK_1a
input = open("input1.txt")
output = open("output1a.txt", "w")

vert_edge = input.readline().strip().split(" ")
vert = int(vert_edge[0])
edges = int(vert_edge[1])

dict1 = {}
for key in range(1, vert+1):
  dict1[key] = []

start = 0
while start < edges:
    line = input.readline().strip().split(" ")
    dict1[int(line[0])].append(int(line[1]))
    start += 1

visited=[]
visiting=[]
def detectCycle(graph, vert):
    global visited
    global visiting
    if vert in visited:
        return False

    elif vert in visiting:
        return True

    visiting.append(vert)

    for adj_vert in graph[vert]:
        if detectCycle(graph, adj_vert):
            return True

    visiting.remove(vert)
    visited.append(vert)

    return False


visited_dfs = []
stack = []

def dfs(graph, vertex):
    global stack
    global visited_dfs
    if vertex not in visited_dfs:
        visited_dfs.append(vertex)

        for adj_vert in graph[vertex]:
                dfs(graph, adj_vert)
        stack.append(vertex)

def TopSort(graph):
    global visited_dfs
    global stack

    for i in range(1, vert+1):

            if detectCycle(graph, i) == True:
                stack = []
                break
            else:
                dfs(graph, i)

    stack.reverse()
    return stack

result = TopSort(dict1)
if result != []:
      for i in result:
        output.write(f"{i} ")
else:
  output.write("IMPOSSIBLE")

output.close()
```

```python
#TASK_1b
input = open("input1.txt")
output = open("output1b.txt", "w")

vert_edge = input.readline().strip().split(" ")
vert = int(vert_edge[0])
edges = int(vert_edge[1])

dict1 = {}
for key in range(1, vert+1):
  dict1[key] = []

start = 0
while start < edges:
    line = input.readline().strip().split(" ")
    dict1[int(line[0])].append(int(line[1]))
    start += 1
indegree = [0] * (vert+1)
for i in range(1, vert+1):
  if dict1[i] != []:
    for j in dict1[i]:
      indegree[j] += 1

def top_sort_kahns(graph):
  count = 0
  visited = []
  queue = []
  result = []
  for i in range(1, vert+1):
    if indegree[i] == 0:
      queue.append(i)

  while queue != []:
    vertex = queue.pop(0)
    result.append(vertex)

    for adj_vert in graph[vertex]:
      if adj_vert not in visited:
          indegree[adj_vert] -= 1
          if indegree[adj_vert] == 0:
            queue.append(adj_vert)
            visited.append(adj_vert)
    count += 1
  return result, count

top_sort, count = top_sort_kahns(dict1)
if count != vert:
    output.write("IMPOSSIBLE")
else:
    for i in top_sort:
        output.write(f"{i} ")

output.close()
```

```python
#TASK_2
import heapq
input = open("input2.txt")
output = open("output2.txt", "w")

vert_edge = input.readline().strip().split(" ")
vert = int(vert_edge[0])
edges = int(vert_edge[1])

dict1 = {}
for key in range(1, vert+1):
  dict1[key] = []

start = 0
while start < edges:
    line = input.readline().strip().split(" ")
    dict1[int(line[0])].append(int(line[1]))
    start += 1
indegree = [0] * (vert+1)
for i in range(1, vert+1):
  if dict1[i] != []:
    for j in dict1[i]:
      indegree[j] += 1

def top_sort_kahns(graph):
  count = 0
  visited = []
  queue = []
  result = []
  for i in range(1, vert+1):
    if indegree[i] == 0:
      queue.append(i)

  while queue != []:
    heapq.heapify(queue)
    vertex = heapq.heappop(queue)
    result.append(vertex)

    for adj_vert in graph[vertex]:
      if adj_vert not in visited:
          indegree[adj_vert] -= 1
          if indegree[adj_vert] == 0:
            queue.append(adj_vert)
            visited.append(adj_vert)
    count += 1
  return result, count

top_sort, count = top_sort_kahns(dict1)
if count != vert:
    output.write("IMPOSSIBLE")
else:
    for i in top_sort:
        output.write(f"{i} ")

output.close()
```
